## Modelo SEIR

El modelo SEIR considera la dinámica de una epidemia dividiendo la población afectada en cuatro grupos: 

1 Los susceptibles (S: susceptible) son aquellos que podrían contraer la enfermedad al estar en contacto con los infectados.

2 Los expuestos (E: exposed) son los que se encuentran en periodo de incubación y por tanto no presentan síntomas, por lo que no transmitirían la infección.

3 Los infectados (I: infected) son los que presentan síntomas y transmiten la enfermedad.

4 Los recuperados o removidos (R: recupered) son los que han sido retirados de la dinámica del estudio y  no vuelven a contagiarse de acuerdo con este modelo.


Este es un modelo de ecuaciomes diferenciales ordinarias, descrito de la siguiente manera:

$$ \frac{dS}{dt} = - \beta SI  $$\
$$ \frac{dE}{dt} = \beta SI - \alpha E  $$\
$$ \frac{dI}{dt} = \alpha E - \gamma I  $$\
$$ \frac{dR}{dt} = \gamma I  $$\
$$ N = S + E + I + R = 1.0 $$

Aquí, las variables de estado de cada modelo son las siguientes:

$S$ Susceptible; Población vulnerable.

$E$ Expuesto; Población expuesta.

$I$ Infectado; Número de población infectada.

$R$ Resistente; Población introvertida.

$N$ Número total de la población.

Los siguientes parámetros dependen del agente infeccioso:

$\alpha$ Porcentaje de personas expuestas.

$\beta$  Este parámetro controla con qué frecuencia los contactos susceptibles causan una nueva exposición.

$\gamma$ La tasa a la cual los pacientes infectados de recuperan y pasan a la etapa de resistencia.

$\mu$ Mortalidad natural (independiente de la enfermedad) Modela una población de cierto tamaño.

$\nu$ Tasa de vacunación en poblaciones vulnerables.


Para la simulación real, las ecuaciones diferenciales deben analizarse numéricamente y deben establecerse los valores iniciales para cada variable de estado y los parámetros apropiados para el covid-19, dependiendo el modelo que se quiere simular, por ejemplo una ciudad, un país etc.

## Modelo SEIR simplificado

En el modelo SEIR descrito anteriormente, si se establece la tasa de mortalidad natural de $\mu$  y la tasa de vacuna  $\nu$ a 0 dará en términos generaleas la misma fórmula. Además, expresar todas las ecuaciones como una fracción de la población da como resultado $N = 1$. 

$$\frac{dS}{dt} = - \beta SI  \tag{1}$$

$$\frac{dE}{dt} = \beta SI- \alpha E  \tag{2}$$

$$\frac{dI}{dt} = \alpha E- \gamma I  \tag{3}$$

$$\frac{dR}{dt} = \gamma I \tag{4}$$

$$ N = S + E + I + R = 1.0 \tag{5}$$


Aquí:

$\alpha$ Es el inverso del período de incubación ($\frac{1}{t_ {incubación}}$).

$\beta$ Es la velocidad de contacto promedio de la población.

$\gamma$ Es el recíproco de la duración promedio de la infección ($\frac{1}{t_ {infeccioso}}$).

La ecuación (1) es el cambio en una persona susceptible a la enfermedad, y está regulada por el número de personas infectadas y el contacto con la persona infectada. La ecuación (2) se da a las personas expuestas a la enfermedad,  crece con la velocidad de contacto y disminuye con el período de incubación, infectando a las personas. La ecuación (3) muestra el cambio en las personas infectadas según la población expuesta y el período de incubación, disminuye con la duración de la infección, por lo que cuanto mayor sea el $\gamma$ más rápido morirá o se recuperará la gente y pasará a la etapa final de la ecuación (4). La ecuación final, número (5), es una restricción que indica que el modelo no tiene efecto de nacimiento / migración. Tenemos una población fija de principio a fin.

Entre los parámetros que se requieren se encuentra el $R_0$. Este valor expresa la rapidez con que la enfermedad se propaga por la ecuación:

$$R_0 = \frac{\beta}{\gamma} \label{eq:R0} \tag{6} $$

$R_0$ es un término matemático que indica cuán contagiosa es una enfermedad infecciosa. También se conoce como el número de reproducción. A medida que una infección se propaga a nuevas personas, se reproduce a sí misma.

$R_0$ le indica el número promedio de personas que contraerán una enfermedad de una persona contagiosa. Se aplica específicamente a una población de personas que previamente estaban libres de infección y que no han sido vacunadas. Si una enfermedad tiene un $R_0$ de 18, una persona que la tiene la transmitirá a un promedio de otras 18 personas, siempre y cuando nadie haya sido vacunado contra ella o ya sea inmune a ella en su comunidad.

Existen tres posibilidades para la propagación o disminución potencial de una enfermedad, dependiendo de su valor $R_0$:

1  Si $R_0 < 1$, cada infección existente causa menos de una infección nueva. En este caso, la enfermedad disminuirá y finalmente desaparecerá.

2  Si $R_0 = 1$, cada infección existente causa una nueva infección. La enfermedad se mantendrá viva y estable, pero no habrá un brote ni una epidemia.

3 Si $R_0 > 1$, cada infección existente causa más de una infección nueva. La enfermedad se propagará entre las personas y puede haber un brote o epidemia.

## Modelo que refleja la distancia social

La distancia social se considera una actividad social importante para prevenir reuniones a gran escala, contacto físico y la propagación de enfermedades transmisibles. Según el modelo anterior, el parámetro que afectarán estas actividades sociales es la tasa de contacto $\beta$.

Hubbs (https://towardsdatascience.com/social-distancing-to-slow-the-coronavirus-768292f04296) introdujo un nuevo parámetro $\rho$ para expresar la distancia social. Suponiendo que $\rho$ tiene un valor constante entre 0 y 1, si $\rho = 0$  todos están aislados, y si $\rho = 1$ no se establece una distancia social.


Al aplicar $\rho$ a las ecuaciones (1) y (2) para el modelo se tiene:

$$ \frac{dS}{dt} = - \rho \beta SI  $$

$$ \frac{dE}{dt} = \rho \beta SI- \alpha E $$

## Implementación del modelo SEIR

El código es tomado de https://towardsdatascience.com/social-distancing-to-slow-the-coronavirus-768292f04296 aplicando el método de Euler semi-implicito

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker

"""
BASE SEIR model
"""
def base_seir_model(init_vals, params, t):
    S_0, E_0, I_0, R_0 = init_vals
    S, E, I, R = [S_0], [E_0], [I_0], [R_0]
    alpha, beta, gamma = params
    dt = t[1] - t[0]
    for _ in t[1:]:
        next_S = S[-1] - (beta*S[-1]*I[-1])*dt
        next_E = E[-1] + (beta*S[-1]*I[-1] - alpha*E[-1])*dt
        next_I = I[-1] + (alpha*E[-1] - gamma*I[-1])*dt
        next_R = R[-1] + (gamma*I[-1])*dt
        S.append(next_S)
        E.append(next_E)
        I.append(next_I)
        R.append(next_R)
    return np.stack([S, E, I, R]).T

"""
SEIR model with social distancing
"""
def seir_model_with_soc_dist(init_vals, params, t):
    S_0, E_0, I_0, R_0 = init_vals
    S, E, I, R = [S_0], [E_0], [I_0], [R_0]
    alpha, beta, gamma, rho = params
    dt = t[1] - t[0]
    for _ in t[1:]:
        next_S = S[-1] - (rho*beta*S[-1]*I[-1])*dt
        next_E = E[-1] + (rho*beta*S[-1]*I[-1] - alpha*E[-1])*dt
        next_I = I[-1] + (alpha*E[-1] - gamma*I[-1])*dt
        next_R = R[-1] + (gamma*I[-1])*dt
        S.append(next_S)
        E.append(next_E)
        I.append(next_I)
        R.append(next_R)
    return np.stack([S, E, I, R]).T
```

#### El efecto de la distancia social.

En comparación con el modelo Base, identificamos el impacto en la exposición y la proporción de poblaciones infectadas y qué tan tarde es el número máximo de personas infectadas.

```python
r_E = results_soc[:,1].max() / results_base[:,1].max()
r_I = results_soc[:,2].max() / results_base[:,2].max()
d_I = ( results_soc[:,2].argmax() - results_base[:,2].argmax() ) / scale_x

print(r'Efecto del distanciamiento social: $\rho={}$'.format(rho))
print(r'  Radio de caida expuesto  = {}'.format(1-r_E))
print(r'  Radio de caida Infectados = {}'.format(1-r_I))
print(r'  Retrazo del pico de infectados = {}'.format(d_I))
```

Una distancia social de $\rho$ = 0.8

El distanciamiento social se calcula para reducir la población expuesta y las personas infectadas en un 23% y 22% respectivamente, y retrasar el día en que la persona infectada alcanza su máximo.

Según los resultados de la simulación, incluso si la distancia social es de aproximadamente el 20% (𝑟ℎ𝑜 = 0.8
) tiene éxito, se calcula que habrá un efecto significativo en la cuarentena. Sin embargo, la distancia social del 20% no se establece porque es necesario practicar la distancia social durante más de 100 días.
